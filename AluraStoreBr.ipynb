#1. Análise do faturamento


```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import pandas as pd

# Nomes das lojas
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Cálculo do faturamento total por loja
faturamento_total = loja['Preço'].sum()
faturamento_total2 = loja2['Preço'].sum()
faturamento_total3 = loja3['Preço'].sum()
faturamento_total4 = loja4['Preço'].sum()

# Lista com os valores
valores_faturamento = [faturamento_total, faturamento_total2, faturamento_total3, faturamento_total4]

# Criação do gráfico
plt.figure(figsize=(10, 6))
barras = plt.bar(nomes_lojas, valores_faturamento, color='skyblue')

# Adiciona o valor de faturamento acima de cada barra
for i, valor in enumerate(valores_faturamento):
    plt.text(i, valor + 20000, f'R${valor:,.2f}', ha='center', va='bottom', fontsize=10)

# Título e rótulos dos eixos
plt.title('Faturamento Total por Loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')

# Formatar eixo Y como moeda
formatter = mtick.FuncFormatter(lambda x, _: f'R${x:,.0f}')
plt.gca().yaxis.set_major_formatter(formatter)

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
import matplotlib.pyplot as plt

def plot_top_bottom_produtos(loja_df, nome_loja, cor_top='green', cor_bottom='red'):
    top = loja_df['Produto'].value_counts().head(10)
    bottom = loja_df['Produto'].value_counts().tail(10)



    top.plot(kind='bar', figsize=(10,6), title=f'Top 10 Produtos Mais Vendidos - {nome_loja}', color=cor_top)
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    bottom.plot(kind='bar', figsize=(10,6), title=f'10 Produtos Menos Vendidos - {nome_loja}', color=cor_bottom)
    plt.xlabel('Produto')
    plt.ylabel('Quantidade Vendida')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Exemplo de uso:
plot_top_bottom_produtos(loja, 'Loja 1', 'green', 'red')
plot_top_bottom_produtos(loja2, 'Loja 2', 'blue', 'yellow')
plot_top_bottom_produtos(loja3, 'Loja 3', 'green', 'red')
plot_top_bottom_produtos(loja4, 'Loja 4', 'blue', 'yellow')


```

```python
def categoria_mais_vendida(loja_df, nome_loja):
    if 'Categoria do Produto' not in loja_df.columns or 'Produto' not in loja_df.columns:
        print(f"❌ Dados da {nome_loja} não contêm a coluna 'Categoria' ou 'Produto'")
        return

    # Contar vendas por produto
    vendas_por_produto = loja_df['Produto'].value_counts()

    # Juntar com as categorias: cria um DataFrame com Produto e Categoria
    df_prod_cat = loja_df[['Produto', 'Categoria do Produto']].drop_duplicates()
    df_prod_cat = df_prod_cat.set_index('Produto')

    # Associa cada produto à sua categoria e soma as vendas
    vendas_por_categoria = vendas_por_produto.groupby(df_prod_cat['Categoria do Produto']).sum()

    # Ordena e pega a categoria mais vendida
    categoria_top = vendas_por_categoria.sort_values(ascending=False).head(1)

    print(f"\n")
    print(f"\n🏆 Categoria mais vendida da {nome_loja}:")
    for cat, qtd in categoria_top.items():
        print(f"🟢 {cat} com {qtd} unidades vendidas")

    # Plotando todas as categorias
    vendas_por_categoria.sort_values().plot(kind='barh', figsize=(10,6), color='purple')
    plt.xlabel('Quantidade Vendida')
    plt.ylabel('Categoria')
    plt.title(f'Vendas por Categoria - {nome_loja}')
    plt.tight_layout()
    plt.show()
```

```python
categoria_mais_vendida(loja, 'Loja 1')
categoria_mais_vendida(loja2, 'Loja 2')
categoria_mais_vendida(loja3, 'Loja 3')
categoria_mais_vendida(loja4, 'Loja 4')
```

```python
def grafico_pizza_categoria(loja_df, nome_loja):
    vendas_por_produto = loja_df['Produto'].value_counts()
    df_prod_cat = loja_df[['Produto', 'Categoria do Produto']].drop_duplicates().set_index('Produto')
    vendas_por_categoria = vendas_por_produto.groupby(df_prod_cat['Categoria do Produto']).sum()

    plt.figure(figsize=(8,8))
    vendas_por_categoria.plot(kind='pie', autopct='%1.1f%%', startangle=90, shadow=True)
    plt.title(f'Distribuição de Vendas por Categoria - {nome_loja}')
    plt.ylabel('')  # remove o label do eixo y
    plt.tight_layout()
    plt.show()

    return vendas_por_categoria
grafico_pizza_categoria(loja, 'Loja 1')
grafico_pizza_categoria(loja2, 'Loja 2')
grafico_pizza_categoria(loja3, 'Loja 3')
grafico_pizza_categoria(loja4, 'Loja 4')
```
